# Data Visualization


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import os
```

```python
df = pd.read_csv('cleaned_startup_funding.csv') 
os.makedirs("images/charts", exist_ok=True)
```

## 1. Startup Fundings Over the Years

```python
funding_per_year=df.groupby('Year').size()
plt.Figure(figsize=(10,5))
sns.lineplot(x=funding_per_year.index,y=funding_per_year.values)
plt.title('funding_per_year')
plt.ylabel('Number of fundings')
plt.xlabel('year')
plt.grid(True)
plt.savefig("images/charts/plot_1.png", bbox_inches='tight')

```

## 2. Total Funding Amount per Year

```python
funding_per_year=df.groupby('Year')['Amount in USD'].sum()
plt.Figure(figsize=(10,5))
sns.barplot(x=funding_per_year.index,y=funding_per_year.values)
plt.title('total funding ammount per year')
plt.ylabel('Total amount per year')
plt.xlabel('Year')
plt.savefig("images/charts/plot_2.png", bbox_inches='tight')
```

## 3. Top 10 Funded Startups

```python
top_startups=df.groupby('Startup Name')['Amount in USD'].sum().nlargest(10)
plt.Figure(figsize=(10,6))
sns.barplot(x=top_startups.index,y=top_startups.values)
plt.title('top_startups')
plt.xlabel('Total Funding(USD)')
plt.savefig("images/charts/plot_3.png", bbox_inches='tight')
```

## 4. Average Funding Per City

```python
avg_funding=df.groupby('City Location')['Amount in USD'].mean().nlargest(10)
plt.Figure(figsize=(10,5))
sns.barplot(x=avg_funding.index,y=avg_funding.values)
plt.title('Average Funding Per Year')
plt.xlabel('Average Amount(USD)')
plt.savefig("images/charts/plot_4.png", bbox_inches='tight')
```

## 5. Investment Type Distribution

```python
top_investments = df['InvestmentType'].value_counts().nlargest(5)
explode = [0.05] * len(top_investments)

plt.figure(figsize=(8, 6))
top_investments.plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=140,
    explode=explode
)
plt.savefig("images/charts/plot_5.png", bbox_inches='tight')
```

## 6. Top 5 Active Investors

```python
top_investors = df['Investors Name'].value_counts().nlargest(5)

plt.figure(figsize=(8, 5))
sns.barplot(x=top_investors.values, y=top_investors.index, palette='Set2',legend=False)
plt.title('Top 5 Most Active Investors')
plt.xlabel('Number of Investments')
plt.savefig("images/charts/plot_5.png", bbox_inches='tight')

```
